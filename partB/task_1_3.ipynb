# Task 1.3 — What the Paper Claims to Improve

## Baseline Method

The primary baseline used in the paper is **Support Vector Machine (SVM) classification with binary unigram feature weighting**. In this representation, each word feature receives a value of 1 if the word appears in the document and 0 otherwise.

This baseline was originally shown to perform well in sentiment classification tasks such as movie reviews and has been widely used in prior work.

**Paper Reference:** Introduction and Section 3 — discussion of binary unigram representation.

---

## Limitation of the Baseline

The limitation identified by the authors is that **binary feature weighting ignores the relative importance of words across documents and across classes**.  

While binary features indicate whether a word is present, they do not consider:

- how frequently the word appears in the document  
- how informative the word is across the entire corpus  
- whether the word occurs more often in one sentiment class than another

Because of this, the representation may fail to highlight words that strongly differentiate **positive and negative sentiment classes**.

**Paper Reference:** Section 3.1 — discussion of term weighting and limitations of binary features.

---

## How the Proposed Method Addresses the Limitation

The paper proposes **Delta TF-IDF and related weighting schemes** adapted from Information Retrieval.

The key idea is to assign **higher weights to words that appear more frequently in one sentiment class than the other**.

This is achieved by computing class-specific document frequencies and modifying the TF-IDF formula to incorporate the difference between the two classes:

$$
w_i = tfi \cdot \log_2
\left(
\frac{N_1 \cdot df_{i,2}}
{df_{i,1} \cdot N_2}
\right)
$$

This weighting scheme highlights words that are strong indicators of sentiment polarity, allowing the classifier to better distinguish positive and negative documents.

**Paper Reference:** Equation (2) and Section 3.2 — Delta TF-IDF formulation.

---

## Scenario Where the Proposed Method May Not Outperform the Baseline

The proposed method may not outperform the binary baseline when **the vocabulary distribution between sentiment classes is very similar**.

For example, in datasets where:

- documents are mostly neutral or objective
- sentiment is expressed indirectly through context rather than specific words
- both classes share nearly identical word distributions

the difference in class-specific document frequencies becomes very small. In such cases, Delta TF-IDF cannot effectively identify discriminative features, and the simpler binary unigram representation may perform similarly or even better.

This limitation arises because the proposed method relies heavily on **differences in word distributions between sentiment classes**.
