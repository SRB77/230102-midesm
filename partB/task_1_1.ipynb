# Step-by-Step Method Description

## Step 1: Problem Definition

The paper studies the problem of **sentiment classification**, where the goal is to determine whether a document expresses **positive or negative sentiment**.

Earlier research showed that using **binary unigram features with an SVM classifier** provides strong baseline performance for sentiment analysis.

However, binary representations ignore the **relative importance of words**, which may limit classification accuracy.

**Reference:** Introduction section of the paper.

**Purpose**

To motivate the need for improved **term weighting schemes** that can better represent sentiment-related words.

---

## Step 2: Bag-of-Words Document Representation

Each document is represented using a **bag-of-words feature vector**, where each unique term in the vocabulary corresponds to one feature dimension.

A document representation is defined as:

$$
D = \{w_1, w_2, ..., w_{|V|}\}
$$

where:

- $|V|$ is the vocabulary size  
- $w_i$ is the weight of term $i$

**Reference:** Section 3 — Document representation.

**Purpose**

This step converts textual documents into numerical vectors that can be processed by machine learning classifiers.

---

## Step 3: Term Frequency (TF)

Term frequency measures how often a word appears in a document.

The classical term frequency definition is:

$$
TF_i = tfi
$$

where $tfi$ represents the number of occurrences of term $i$ in document $D$.

The paper also studies **sublinear TF scaling**, such as logarithmic scaling:

$$
TF_i = 1 + \log(tf_i)
$$

which prevents repeated terms from dominating the representation.

**Reference:** Table 1 (SMART notation for TF variants).

**Purpose**

To capture the **importance of a word within a document** while preventing excessive influence from repeated terms.

---

## Step 4: Inverse Document Frequency (IDF)

Inverse Document Frequency measures how rare a term is across the entire dataset.

The classic TF-IDF formulation is:

$$
w_i = tfi \cdot idfi
$$

where:

$$
idfi = \log \left(\frac{N}{df_i}\right)
$$

- $N$ = total number of documents  
- $df_i$ = number of documents containing term $i$

**Reference:** Equation (1) in Section 3.1.

**Purpose**

To reduce the weight of very common words while emphasizing **informative terms**.

---

## Step 5: Delta TF-IDF Weighting

The main contribution of the paper is **Delta TF-IDF**, which modifies TF-IDF for classification tasks.

Instead of using global document frequency, the method computes **class-specific document frequencies**.

The Delta TF-IDF weight is defined as:

$$
w_i = tfi \cdot \log_2
\left(
\frac{N_1 \cdot df_{i,2}}
{df_{i,1} \cdot N_2}
\right)
$$

where:

- $N_1$ = number of documents in class 1  
- $N_2$ = number of documents in class 2  
- $df_{i,1}$ = number of documents in class 1 containing term $i$  
- $df_{i,2}$ = number of documents in class 2 containing term $i$

**Reference:** Equation (2), Section 3.2.

**Purpose**

To emphasize words that strongly distinguish **positive and negative sentiment classes**.

---

## Step 6: SMART and BM25 Weighting Variants

The paper also evaluates advanced weighting schemes derived from **Information Retrieval systems**, including:

- SMART weighting variants  
- BM25 weighting functions

These schemes incorporate **sublinear term frequency scaling**, document length normalization, and alternative inverse document frequency formulations.

**Reference:** Section 3.3 and Tables 1–3.

**Purpose**

To test whether sophisticated weighting schemes used in **search engines** can improve sentiment classification accuracy.

---

## Step 7: Sentiment Classification using SVM

The weighted document vectors are used to train a **Support Vector Machine (SVM)** classifier.

SVM learns a decision boundary separating **positive and negative sentiment documents**.

**Reference:** Section 4 (Experimental Setup).

**Purpose**

To evaluate how different weighting schemes affect classification performance.

---

## Step 8: Evaluation on Multiple Sentiment Datasets

The proposed weighting schemes are tested on several datasets, including:

- Movie Review dataset (Pang et al. 2002)
- Multi-Domain Sentiment dataset (Amazon reviews)
- BLOGS06 dataset

Performance is measured using **classification accuracy**.

**Reference:** Section 4 and Table 5.

**Purpose**

To empirically evaluate whether the proposed weighting schemes improve sentiment classification performance.

---

# Final Summary

This paper addresses the problem of improving sentiment classification by studying advanced **term weighting schemes derived from Information Retrieval**. The authors show that **Delta TF-IDF and smoothed weighting variants significantly improve accuracy compared to traditional binary unigram representations**, particularly when combined with sublinear TF scaling and document frequency smoothing.
