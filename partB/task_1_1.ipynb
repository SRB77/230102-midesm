# Step-by-Step Method Description

## Step 1: Problem Definition

The paper addresses the task of **sentiment classification**, where the objective is to automatically determine whether a text document expresses **positive or negative sentiment**.

Most previous approaches represent documents using **binary unigram features**, where each word is represented simply by its presence or absence in the document. While this representation works well in practice, it ignores differences in **word importance** across documents.

**Purpose**

This step defines the research problem: improving document representation so that sentiment classifiers can more effectively distinguish positive and negative opinions.

---

## Step 2: Document Representation using Bag-of-Words

Each document is transformed into a **bag-of-words feature vector** where every unique word in the vocabulary corresponds to one feature dimension.

The document representation can be written as:

$$
D = \{w_1, w_2, ..., w_{|V|}\}
$$

where:

- $|V|$ is the vocabulary size  
- $w_i$ represents the weight assigned to word $i$

**Purpose**

This step converts raw text into a numerical representation suitable for machine learning algorithms.

---

## Step 3: Term Frequency Computation (TF)

For each word in a document, the **term frequency (TF)** is computed to measure how often that word appears in the document.

$$
TF_i = \text{number of occurrences of word } i \text{ in document } D
$$

The paper also evaluates **sublinear TF scaling**, where the frequency is transformed using a logarithmic function to prevent repeated words from dominating the representation.

Example of sublinear scaling:

$$
TF_i = 1 + \log(\text{count}_i)
$$

**Purpose**

TF captures the **local importance of a word within a document**.

---

## Step 4: Inverse Document Frequency (IDF)

To measure how informative a word is across the entire dataset, the method computes **inverse document frequency (IDF)**:

$$
IDF_i = \log \left(\frac{N}{DF_i}\right)
$$

where:

- $N$ = total number of documents  
- $DF_i$ = number of documents containing word $i$

Words that appear in many documents receive **lower weight**, while rare words receive **higher weight**.

**Purpose**

IDF reduces the importance of common words while emphasizing **discriminative terms**.

---

## Step 5: Delta TF-IDF Weighting

The main contribution of the paper is the introduction of **Delta TF-IDF**, which adapts TF-IDF specifically for **classification tasks**.

Instead of computing document frequency globally, the method calculates frequencies separately for **positive and negative classes**.

The weighting function becomes:

$$
w_i = TF_i \cdot
\log_2 \left(
\frac{N_1 \cdot DF_{i,2}}
{N_2 \cdot DF_{i,1}}
\right)
$$

where:

- $N_1$ = number of positive documents  
- $N_2$ = number of negative documents  
- $DF_{i,1}$ = document frequency of word $i$ in positive documents  
- $DF_{i,2}$ = document frequency of word $i$ in negative documents  

**Purpose**

This step assigns higher weights to words that strongly differentiate **positive and negative sentiment**.

---

## Step 6: Feature Vector Construction

After computing the weights, each document is represented as a **weighted feature vector**.

The paper evaluates several weighting schemes including:

- Binary weighting  
- TF-IDF  
- SMART weighting variants  
- BM25 weighting  
- Delta TF-IDF variants

**Purpose**

To compare different weighting schemes and determine which representation produces the best sentiment classification performance.

---

## Step 7: Sentiment Classification using SVM

The weighted feature vectors are used to train a **Support Vector Machine (SVM)** classifier.

SVM learns a decision boundary that separates **positive and negative documents** in the feature space.

**Purpose**

To evaluate whether improved weighting schemes lead to better sentiment classification accuracy.

---

## Step 8: Evaluation on Multiple Datasets

The method is tested on several sentiment datasets, including:

- Movie review dataset  
- Amazon product review dataset  
- Blog opinion dataset  

Performance is measured using **classification accuracy**.

**Purpose**

To empirically validate whether the proposed weighting schemes outperform traditional binary representations.

---

# Final Summary

This paper addresses the problem of improving **sentiment classification through better document representation**. The authors propose **Delta TF-IDF and advanced weighting schemes that incorporate class-specific document statistics**, allowing sentiment-discriminative words to receive higher importance compared to traditional binary or standard TF-IDF representations.
