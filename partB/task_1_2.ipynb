# Task 1.2 — Key Assumptions

## Assumption 1

**Assumption:**  
The method assumes that **words expressing sentiment occur more frequently in one class (positive or negative) than the other**.

**Why the method needs it:**  
The proposed Delta TF-IDF weighting scheme computes weights based on the difference between class-specific document frequencies. If certain words appear significantly more often in one class, the weighting scheme assigns them higher importance for classification.

**Violation scenario:**  
If both classes contain very similar vocabulary distributions (for example, neutral news articles or objective technical documents), then the difference in document frequencies between classes would be minimal. In such cases, Delta TF-IDF would fail to highlight discriminative features for sentiment classification.

**Paper reference:**  
Section 3.2 — Delta TF-IDF formulation.

## Assumption 2

**Assumption:**  
The method assumes that **the importance of a term does not increase linearly with its frequency in a document**.

**Why the method needs it:**  
The paper introduces **sublinear term frequency scaling** (e.g., logarithmic TF) because repeated occurrences of a word should not proportionally increase its importance. This assumption is used to prevent very frequent terms from dominating the document representation.

**Violation scenario:**  
In domains where repeated word occurrences actually indicate stronger sentiment or importance (for example, social media posts with repeated emphasis such as “very very very good”), sublinear scaling might underestimate the importance of those terms.

**Paper reference:**  
Section 3.3 and Table 1 — SMART term-frequency weighting variants.

---

## Assumption 3

**Assumption:**  
The method assumes that **document frequency values must be smoothed to avoid instability when a term appears in only one class**.

**Why the method needs it:**  
Without smoothing, the document frequency of a term in one class could be zero, which would cause division-by-zero errors in the Delta TF-IDF formula. Smoothing ensures stable weight computation even when terms occur exclusively in one class.

**Violation scenario:**  
In very small datasets where smoothing significantly alters the weight calculations, the smoothing factor might introduce bias and distort the importance of certain terms.

**Paper reference:**  
Section 3.4 — Smoothed Delta TF-IDF variants.

---

## Assumption 4

**Assumption:**  
The method assumes that **bag-of-words representation is sufficient to capture sentiment information in documents**.

**Why the method needs it:**  
The proposed approach operates entirely on word-level features and does not incorporate syntactic structure or contextual information. Therefore, it assumes that the presence and weighting of individual words are enough to determine sentiment polarity.

**Violation scenario:**  
In cases where sentiment depends heavily on context or compositional meaning (for example sarcasm or negation such as “not good”), bag-of-words representations may fail to capture the true sentiment.

**Paper reference:**  
Section 3 — Document representation using bag-of-words.
